```python
import pandas as pd
import numpy as np

# Sample data
data = {'v1': np.random.randint(1, 1000, 10)}  # 10 random integers between 1 and 1000
df = pd.DataFrame(data)

# Use qcut to split into 3 groups based on quantiles
df['v2'] = pd.qcut(df['v1'], q=3, labels=['group1', 'group2', 'group3'])

print(df)
```

```python
# Use cut to segment values into specific bins
df['v2'] = pd.cut(df['v1'], bins=[1, 100, 250, np.inf], labels=['group1', 'group2', 'group3'])

print(df)
```
